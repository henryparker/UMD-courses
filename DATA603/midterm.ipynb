8. Entropy and Gini Impurity

```python
from scipy.stats import entropy
entropy([1/3, 1/3, 1/3], base=2)
```

```python
entropy([1/3, 2/3], base=2) - 2/3 * entropy([1/2, 1/2], base=2)
```

```python
def Gini_Impurity(data):
    total_weight = sum(data)
    return 1 - sum([(v/total_weight)**2 for v in data])

Gini_Impurity([2, 3])
```
