This dataset contains information collected through a survey conducted among students. It includes various attributes related to their preferences, interests, and social behavior. Each entry in the dataset represents a student, and the attributes captured are as follows:


* **Timestamp:** The timestamp indicating when the survey was completed.

* **Score:** The score achieved by the student (if applicable) [WE IGNORED IT].
* **Name:** The name of the student.

* **Age:** The age of the student.
* **Start_Date:** The approximate date when the student started their first class at UMD.
* **Major:** The major or field of study of the student.
* **Favorite_Color:** The student's most favorite color.
* **Favorite_Hobby:** The student's most favorite hobby, provided as a single word in all lowercase letters.
* **Favorite_Sports:** The student's most favorite sport, provided as a single word in all lowercase letters.
* **Favorite_Movie:** The student's most favorite movie, provided in all lowercase letters.
* **Favorite_Series:** The student's most favorite series, provided in all lowercase letters.
* **Career_Interest:** The student's career interest or aspiration.
* **Friend_Quality_Preference:** The quality or trait that the student looks for in a friend.
* **Similar_Interest_Preference:** The student's agreement level with the statement "I prefer friends with similar interests".
* **Importance_of_Color_Similarity:** The importance level, on a scale of 1 to 5, for the student to have friends who share their favorite color.
* **Importance_of_Movie_Similarity:** The importance level, on a scale of 1 to 5, for the student to have friends who share their favorite movie.
* **Importance_of_Series_Similarity:** The importance level, on a scale of 1 to 5, for the student to have friends who share their favorite series.
* **Importance_of_Hobby_Similarity:** The importance level, on a scale of 1 to 5, for the student to have friends who share their favorite hobby.
* **Importance_of_Career_Similarity:** The importance level, on a scale of 1 to 5, for the student to have friends who share their career interest.
* **Social_Activities_Frequency:** The frequency of the student's engagement in social activities with friends.
* **Social_Gathering_Preference! :** The student's preference for small group hangouts or larger social gatherings.

* **Outgoing_or_Introverted:** The student's self-identification as more outgoing or introverted.

This dataset provides insights into the preferences, interests, and social behaviors of the surveyed students, facilitating analysis of friendship dynamics and social interactions.

```python
# Function to convert days to seconds
def days_to_seconds(date):
    # here ".days" attribute returns the number of days in the Timedelta object, which represents
    # the number of days between date and January 1, 2018.
    days = (date - pd.Timestamp('2018-01-01')).days
    return days * 24 * 60 * 60

def days(date):
    days = (date - pd.Timestamp('2018-01-01')).days
    return days

# Convert 'Start_Date' column to datetime format
df['Start_Date'] = pd.to_datetime(df['Start_Date'])

# Apply function to 'Start_Date' column
df['date_indays_since_zero'] = df['Start_Date'].apply(days_to_seconds)
df
```

# Outlier Detection using Z-Score

```python
# Function to calculate Z-scores
def z_score(df, threshold=3.5):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    z_scores = np.abs((df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std())
    outliers = (z_scores > threshold).any(axis=1)
    return outliers

# Detect outliers
outliers = z_score(df)

# Show outliers
print("Outliers detected:")
df[outliers]
```
