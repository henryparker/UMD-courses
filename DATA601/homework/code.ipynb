### Q1

```python
S = [1,2,3,4,5,6]
X = [1/2 * (s-3) * (s - 5) for s in S]
print(sorted(X))
# Expectation
prob = np.array([1/6]*6)
rm = np.array(X)
E = np.dot(prob, rm)
print(E)
# Variance
EX2 = np.dot(prob, rm**2)
print(EX2-E**2)

values, counts = np.unique(X, return_counts=True)

pmf = counts / len(X)

plt.xticks(np.arange(-1, 4 + 1, 0.5))

plt.bar(values, pmf, width=0.1)
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('PMF')
plt.savefig('homework5/q1pmf.jpg')

cdf = np.cumsum(pmf)

# 绘制CDF图像
plt.figure()
plt.xticks(np.arange(-1, 4 + 1, 0.5))
plt.plot(values, cdf, marker='o', linestyle='-')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF')
plt.savefig('homework5/q1cdf.jpg')

variance = np.var(X, ddof=0)
print(variance)
```
