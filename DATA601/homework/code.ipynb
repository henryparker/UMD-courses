### Q1

```python
S = [1,2,3,4,5,6]
X = [1/2 * (s-3) * (s - 5) for s in S]
print(sorted(X))
# Expectation
prob = np.array([1/6]*6)
rm = np.array(X)
E = np.dot(prob, rm)
print(E)
# Variance
EX2 = np.dot(prob, rm**2)
print(EX2-E**2)

values, counts = np.unique(X, return_counts=True)

pmf = counts / len(X)

plt.xticks(np.arange(-1, 4 + 1, 0.5))

plt.bar(values, pmf, width=0.1)
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('PMF')
plt.savefig('homework5/q1pmf.jpg')

cdf = np.cumsum(pmf)

# 绘制CDF图像
plt.figure()
plt.xticks(np.arange(-1, 4 + 1, 0.5))
plt.plot(values, cdf, marker='o', linestyle='-')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF')
plt.savefig('homework5/q1cdf.jpg')

variance = np.var(X, ddof=0)
print(variance)
```

```python
import numpy as np

# 定义一个方阵
A = np.array([[1, 0, 0],
              [0, 1, 0],
              [1/36, 1/36, 17/18]])


# 求解特征值和特征向量
eigenvalues, eigenvectors = np.linalg.eig(A)

# 构造对角矩阵和变换矩阵
D = np.diag(eigenvalues)  # 对角矩阵
P = eigenvectors          # 特征向量矩阵

# 验证对角化：A = P * D * P^(-1)
A_reconstructed = np.array([0, 0, 1]) @ P @ np.linalg.matrix_power(D, 1000) @ np.linalg.inv(P)
A_reconstructed
```
